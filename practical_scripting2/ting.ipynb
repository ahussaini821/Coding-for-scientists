```python
from Bio.Seq import Seq
from Bio import SeqIO
print "Program start"

# 16 chromosomes so range of 1,17
for i in range(1,17):
    
    # Current chromosome we are checking
    curr_record = SeqIO.parse("crosses.chr{}.fa".format(i), "fasta")
    
    # Current file we are writing to
    curr_file = open("binary {}".format(i), "w")
    
    # Only need to open DBV as base either matches to DBV, YPS, or is an N
    dbv = open("DBV {}".format(i), "r")
    dbv_content = dbv.readlines()
    
    for record in curr_record:
        
        curr_file.write(record.id + "\n")
        
        # Iteartes through current sequence giving index and base
        for index, base in enumerate(record.seq):
             
            #print base + dbv_content[index].strip("\n")
            if dbv_content[index].strip("\n") == base:
                curr_file.write("1")
            elif base == "N":
                curr_file.write("N")
            else:
                curr_file.write("0")
             
            
        curr_file.write("\n")
    curr_file.close()
    dbv.close()
    
print "Program end"

```
